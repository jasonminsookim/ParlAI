# Import libraries and data

```python
import pandas as pd
import os
import sys

sys.path.append('..')
```

```python
scores = pd.read_csv('../data/scored_chatbot_interactions/raw/JasonMcFadden_scored_chats.csv')
scores.head()
```

```python
scored = scores.dropna().copy()
```

```python
print(scores.shape)
```

```python
print(scored.shape)
```

```python
scored.head(20)
```

```python
def clean_scored_text(text):
    clean_text = str(text).replace('[TransformerGenerator]:', '')
    clean_text = clean_text.replace('Enter Your Message:', '')
    return clean_text
```

```python
scored['input'] = scored['input'].apply(clean_scored_text)
scored['cb_response_1'] = scored['cb_response_1'].apply(clean_scored_text)
scored['input_2'] = scored['input_2'].apply(clean_scored_text)
scored['cb_response_2'] = scored['cb_response_2'].apply(clean_scored_text)
scored['input_3'] = scored['input_3'].apply(clean_scored_text)
scored['cb_response_3'] = scored['cb_response_3'].apply(clean_scored_text)
```

```python
scored.to_csv('../data/scored_chatbot_interactions/processed/JasonMcFadden_scored_chats_processed.csv')
```
